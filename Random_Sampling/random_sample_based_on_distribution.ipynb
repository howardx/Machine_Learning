# standard normal distributed sample of size 1000

```python
std_norm_time_series_sample = np.random.randn(1000, 1) # generate 1000 vectors of 1D
time_axis = range(1000)

print std_norm_time_series_sample.shape
print len(time_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.scatter(time_axis, std_norm_time_series_sample) # X axis - time, Y axis - standard normal sample
plt.show()
```
